```python
from IPython.display import display

import numpy as np
import pandas as pd

Dataset = pd.read_csv('energydata_complete.csv')
display(Dataset.head())
```

```python
# Checking Empty & NaN Value
Dataset.isnull().sum()
Dataset.isna().sum()
```

```python
# Question No. 12
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

a, b, c, d = train_test_split(Dataset[['T2']], Dataset['T6'], test_size=0.3, random_state=2)

l_model = LinearRegression()
l_model.fit(a, c)
pred_values = l_model.predict(b)

# R-Squared Score
from sklearn.metrics import r2_score
r2_score = r2_score(d, pred_values)
round(r2_score, 2) 
```

```python
df_1 = Dataset.drop(columns = ['date', 'lights'], axis=1)

# Normalization
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()

norm_data = pd.DataFrame(Scaler.fit_transform(df_1), columns=df_1.columns)

x_1 = norm_data.drop(columns=['Appliances'])
y_1 = norm_data['Appliances']

# Regression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_1, y_1, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
# Question No. 13
# MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
# Question No. 14
# RSS
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
# Question No. 15
# RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
# Question No. 16
# R-Squared or Coefficient of Determination
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted_values)
round(r2, 2) 
```

```python
# Question No. 17
# Feature Weights

def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
    
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights
```

```python
# Question No. 18
# Ridge Regression

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

Pred_1 = ridge_reg.predict(x_test)

# RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, Pred_1))
round(rmse, 3)
```

```python
# Question No. 19 & 20
# LASSO Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

Pred_2 = lasso_reg.predict(x_test)

# RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, Pred_2))
print(round(rmse, 3))

print(np.sum(lasso_reg.coef_ != 0))
```
